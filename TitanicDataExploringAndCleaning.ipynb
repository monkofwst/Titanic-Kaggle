```python
import numpy as np
import pandas as pd
import keras

train = pd.read_csv("C:/Users/bazar/Downloads/train.csv")
test = pd.read_csv("C:/Users/bazar/Downloads/test.csv")
```

```python
def clear(ds):
    ds.drop('Ticket', axis=1, inplace=True)
    ds.drop('Cabin', axis=1, inplace=True)
    ds.drop('PassengerId', axis=1, inplace=True)
    ds.drop('Name', axis=1, inplace=True)
    cols = ['SibSp', 'Parch', 'Fare', 'Age']
    for col in cols:
        ds[col].fillna(ds[col].median(), inplace=True)
    ds.Embarked.fillna('U', inplace=True)
    cols = ['Sex', 'Embarked']
    for col in cols:
        sl = keras.layers.StringLookup(output_mode='int') 
        sl.adapt(ds[col])
        ds[col] = sl(ds[col])
    cols = ['Age', 'Fare']
    for col in cols:
        ds[col] -= ds[col].mean()
        ds[col] /= ds[col].std()
    return ds
```
