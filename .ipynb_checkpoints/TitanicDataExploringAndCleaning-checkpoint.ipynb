```python
ds = pd.read_csv("C:/Users/bazar/Downloads/test.csv")
ds
```

```python
ds.drop('Ticket', axis=1, inplace=True)
ds.drop('Cabin', axis=1, inplace=True)
ds.drop('Name', axis=1, inplace=True)
ds = ds[ds.Age.notnull()]
ds.Embarked.fillna('U', inplace=True)
import keras
sl_sex = keras.layers.StringLookup(max_tokens=2, output_mode='int')
sl_sex.adapt(ds.Sex)
ds.Sex = sl_sex(ds.Sex)
sl_emb = keras.layers.StringLookup(output_mode='int')
sl_emb.adapt(ds.Embarked)
ds.Embarked = sl_emb(ds.Embarked)
ds.Age -= ds.Age.mean()
ds.Age /= ds.Age.std()
ds.Fare -= ds.Fare.mean()
ds.Fare /= ds.Fare.std()
ds.to_csv('test_clean.csv')
```
